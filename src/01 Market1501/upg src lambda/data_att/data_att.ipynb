```python
# Import Libraries
import pandas as pd
import numpy as np
import os
```

```python
# Database loading
global train
global test
train     = '../data/bounding_box_train'
train_att = '../market_attribute_train.csv'

test      = '../data/bounding_box_test'
test_att  = '../market_attribute_test.csv'
```

```python
import os

# Directory containing the training images
train_dir = '../data/bounding_box_train'

# List all files in the directory
train_imgs = os.listdir(train_dir)

# Display the first 10 images in the list
print(train_imgs[:10])
```

```python
import os
import pandas as pd

train = os.listdir(train_dir)  # This will list all files in the directory

train_imgs_data = []
for s in train:
    # Ensure the file is a .jpg image before processing
    if s.endswith('.jpg'):
        # Keep the ID as a string
        id_str = s.split('_')[0]
        train_imgs_data.append([id_str, s])

# Creating a DataFrame from the list
train_imgs_df = pd.DataFrame(train_imgs_data, columns=['id', 'filename'])

# Print the DataFrame
print(train_imgs_df)
```

```python
# Assuming df_t has already been loaded and the 'image_index' column needs cleaning
df_t['image_index'] = df_t['image_index'].str.extract(r"\['(\d+)'\]")
```

```python
df_t
```

```python
# Rename the 'image_index' column in df_t to 'id'
df_t.rename(columns={'image_index': 'id'}, inplace=True)
```

```python
df_t
```

```python
# Merge df_t with train_imgs_df on 'id'
merged_df = pd.merge(df_t, train_imgs_df, on='id', how='left')
```

```python
merged_df
```

```python
# Verify the merge by checking the first few rows
print(merged_df.head())

# Check if there are any rows where 'filename' might be NaN to ensure all ids had a matching filename
print(merged_df[merged_df['filename'].isna()])

# Optionally, save the merged DataFrame to a CSV file for further use
merged_df.to_csv('../train.csv', index=False)
```

```python
import os

# Directory containing the training images
test_dir = '../data/bounding_box_test'

# List all files in the directory
test_imgs = os.listdir(test_dir)

# Display the first 10 images in the list
print(test_imgs[:10])
```

```python
import os
import pandas as pd

test = os.listdir(test_dir)  # This lists all files in the directory

test_imgs_data = []
for s in test:
    if s.endswith('.jpg'):
        # Split and convert ID, replacing -1 with 1
        id_str = s.split('_')[0]
        id_num = int(id_str) if id_str != '-1' else 1  # Convert to integer, replace -1 with 1
        
        # Format the ID to have leading zeros, ensuring 4 digits
        formatted_id = f"{id_num:04d}"
        
        # Append formatted ID and filename to the list
        test_imgs_data.append([formatted_id, s])

# Creating a DataFrame from the list
test_imgs_df = pd.DataFrame(test_imgs_data, columns=['id', 'filename'])

# Print the DataFrame
print(test_imgs_df)
```

```python
# Assuming df_t has already been loaded and the 'image_index' column needs cleaning
df_te['image_index'] = df_te['image_index'].str.extract(r"\['(\d+)'\]")
```

```python
df_te
```

```python
# Rename the 'image_index' column in df_t to 'id'
df_te.rename(columns={'image_index': 'id'}, inplace=True)
```

```python
df_te
```

```python
# Merge df_t with train_imgs_df on 'id'
merged_df_2 = pd.merge(df_te, test_imgs_df, on='id', how='left')
```

```python
# Verify the merge by checking the first few rows
print(merged_df_2.head())

# Check if there are any rows where 'filename' might be NaN to ensure all ids had a matching filename
print(merged_df_2[merged_df_2['filename'].isna()])

# Optionally, save the merged DataFrame to a CSV file for further use
merged_df_2.to_csv('../test.csv', index=False)
```
